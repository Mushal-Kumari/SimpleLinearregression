```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
x=df[['year']]
y=df['per capita income (US$)']
x
```

```python
plt.scatter(x,y)
plt.xlabel("Year")
plt.ylabel("Income")
plt.title("Year vs Canada House Income")
plt.show()
```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
print('y=',reg.intercept_,'+',reg.coef_,'x')
```

```python
df1=pd.read_csv("years.csv")
df1.head()
```

```python
pred=reg.predict(df1)
pred
```

```python
df1['pred']=pred
df1
```

```python
df1.to_csv('prediction.csv')
```
